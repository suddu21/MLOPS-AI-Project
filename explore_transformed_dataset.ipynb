```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")
%matplotlib inline
```

```python
# Load the transformed dataset
# Update the path to match where your transformed_data.csv is saved (e.g., /mnt/c/Users/YourUser/data/transformed_data.csv)
data_path = 'transformed_data.csv'
df = pd.read_csv(data_path)

# Display the first few rows
df.head()
```

```python
# Basic statistics of the dataset
df.describe()
```

```python
# Check the shape of the dataset (rows, columns)
print(f"Dataset shape: {df.shape}")

# Check for missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Count of Class 0 (legitimate) vs Class 1 (fraud)
class_counts = df['Class'].value_counts()
print("\nClass distribution:\n", class_counts)

# Visualize the class distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title('Class Distribution (0: Legitimate, 1: Fraud)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation matrix of features (optional, for deeper exploration)
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, fmt='.2f')
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
# Box plot of Amount by Class (to see distribution differences)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Amount Distribution by Class')
plt.xlabel('Class (0: Legitimate, 1: Fraud)')
plt.ylabel('Amount')
plt.show()
```
